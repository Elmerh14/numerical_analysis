# CP.1.1.2, Sauer3
Use the Bisection Method to find the root to eight correct decimal places for
(a) $x^5 + x = 1$, (b) $\sin(x) = 6x + 5$, (c) $\ln(x) + x^2 = 3$.


```python
import math
import numpy as np
from scipy.optimize import fsolve
```

```python
def bisection(f, a, b, tol):
    fa = f(a)
    while (b - a) / 2 > tol:
        c = (a + b) / 2
        fc = f(c)
        if fc == 0:
            return c
        elif fa * fc < 0:
            b = c
        else:
            a, fa = c, fc
    return (a + b) / 2

import math

tol = 0.5e-8

root_a = bisection(lambda x: x**5 + x - 1,        0, 1, tol)
root_b = bisection(lambda x: math.sin(x) - 6*x - 5, -1, 0, tol)
root_c = bisection(lambda x: math.log(x) + x**2 - 3, 1, 2, tol)

print(f"(a) root = {root_a:.8f}")
print(f"(b) root = {root_b:.8f}")
print(f"(c) root = {root_c:.8f}")
```

# CP.1.2.4, Sauer3
Calculate the cube roots of the following numbers to eight correct decimal places, by using Fixed-Point Iteration with $g(x) = \dfrac{2x + A/x^2}{3}$, where $A$ is (a) 2, (b) 3, (c) 5. State your initial guess and the number of steps needed.


```python
def fixed_point(g, x0, tol, max_iter=1000):
    x = x0
    for n in range(1, max_iter + 1):
        x_new = g(x)
        if abs(x_new - x) < tol:
            return x_new, n
        x = x_new
    raise RuntimeError("Did not converge within max_iter iterations")

# tolerance for 8 correct decimal places
tol = 0.5e-8

# g(x) = (2x + A/x^2) / 3
def g(x, A):
    return (2*x + A/x**2) / 3

# initial guess
x0 = 1.0

for label, A in [('a', 2), ('b', 3), ('c', 5)]:
    root, steps = fixed_point(lambda x: g(x, A), x0, tol)
    print(f"({label}) A = {A}:  cube root ≈ {root:.8f}   (initial guess x0 = {x0}, steps = {steps})")
```

# CP.1.3.2, Sauer3
Let $f(x) = \sin(x^3) - x^3$.
(a) Find the multiplicity of the root $r = 0$.
(b) Use the `scipy.optimize.fsolve` command with initial guess $x = 0.1$ to locate a root. What are the forward and backward errors of `scipy.optimize.fsolve`'s response?

```python

# (a) Multiplicity of the root r = 0 for f(x) = sin(x^3) - x^3
f = lambda x: np.sin(x**3) - x**3

# f(x) ~ C * x^m near the root, so log|f(x)| ~ m*log|x| + log|C|
# estimate m using two small points via the slope of log|f| vs log|x|
h1 = 0.1
h2 = 0.05
m_estimate = (math.log(abs(f(h1))) - math.log(abs(f(h2)))) / (math.log(h1) - math.log(h2))

print(f"f(0.1)  = {f(h1):.6e}")
print(f"f(0.05) = {f(h2):.6e}")
print(f"Estimated multiplicity m = {m_estimate:.4f}")
multiplicity = round(m_estimate)
print(f"Multiplicity of root r = 0 is: {multiplicity}")
print()

# (b) fsolve with initial guess x0 = 0.1, then error analysis
x0 = 0.1
root = fsolve(f, x0)[0]
true_root = 0.0

forward_error = abs(root - true_root)
backward_error = abs(f(root))

print(f"fsolve result:           x = {root:.16e}")
print(f"True root:               r = {true_root}")
print(f"Forward error  |x - r|  = {forward_error:.6e}")
print(f"Backward error |f(x)|   = {backward_error:.6e}")
```

# CP.1.4.2, Sauer3
Each equation has one real root. Use Newton's Method to approximate the root to eight correct decimal places.

a. $x^5 + x = 1$

b. $\sin x = 6x + 5$

c. $\ln x + x^2 = 3$

```python
tol = 0.5e-8
max_iter = 100

#a
f_a  = lambda x: x**5 + x - 1
fp_a = lambda x: 5*x**4 + 1

x = 0.75
print("Part (a): x^5 + x - 1 = 0")
print(f"x0 = {x}")

for n in range(max_iter):
    fx, fpx = f_a(x), fp_a(x)
    x_new = x - fx/fpx
    print(f"  n={n}: x = {x:.10f}, f(x) = {fx:.3e}, x_new = {x_new:.10f}")
    if abs(x_new - x) < tol:
        x = x_new
        break
    x = x_new
root_a = x
print(f"Converged: root = {root_a:.8f} after {n+1} iterations\n")

#b
f_b  = lambda x: np.sin(x) - 6*x - 5
fp_b = lambda x: np.cos(x) - 6

x = -1.0
print("Part (b): sin(x) - 6x - 5 = 0")
print(f"x0 = {x}")

for n in range(max_iter):
    fx, fpx = f_b(x), fp_b(x)
    x_new = x - fx/fpx
    print(f"  n={n}: x = {x:.10f}, f(x) = {fx:.3e}, x_new = {x_new:.10f}")
    if abs(x_new - x) < tol:
        x = x_new
        break
    x = x_new
root_b = x
print(f"Converged: root = {root_b:.8f} after {n+1} iterations\n")

#c
f_c  = lambda x: np.log(x) + x**2 - 3
fp_c = lambda x: 1/x + 2*x

x = 1.5
print("Part (c): ln(x) + x^2 - 3 = 0")
print(f"x0 = {x}")

for n in range(max_iter):
    fx, fpx = f_c(x), fp_c(x)
    x_new = x - fx/fpx
    print(f"  n={n}: x = {x:.10f}, f(x) = {fx:.3e}, x_new = {x_new:.10f}")
    if abs(x_new - x) < tol:
        x = x_new
        break
    x = x_new
root_c = x
print(f"Converged: root = {root_c:.8f} after {n+1} iterations")
```

# CP.1.5.1, Sauer3 Langou
Use the Secant Method to find the (single) solution of each of the following equations. Use initial guesses $x_0 = 1$ and $x_1 = 2$.

a. $x^3 = 2x + 2$

b. $e^x + x = 7$

c. $e^x + \sin x = 4$

```python
tol = 0.5e-8
max_iter = 100

#a
f_a = lambda x: x**3 - 2*x - 2

x0, x1 = 1.0, 2.0
print("Part (a): x^3 - 2x - 2 = 0")
print(f"x0 = {x0}, x1 = {x1}")

for n in range(max_iter):
    f0, f1 = f_a(x0), f_a(x1)
    x2 = x1 - f1*(x1 - x0)/(f1 - f0)
    print(f"  n={n}: x2 = {x2:.10f}, f(x2) = {f_a(x2):.3e}")
    if abs(x2 - x1) < tol:
        x0, x1 = x1, x2
        break
    x0, x1 = x1, x2
root_a = x1
print(f"Converged: root = {root_a:.8f} after {n+1} iterations\n")

#b
f_b = lambda x: np.exp(x) + x - 7

x0, x1 = 1.0, 2.0
print("Part (b): e^x + x - 7 = 0")
print(f"x0 = {x0}, x1 = {x1}")

for n in range(max_iter):
    f0, f1 = f_b(x0), f_b(x1)
    x2 = x1 - f1*(x1 - x0)/(f1 - f0)
    print(f"  n={n}: x2 = {x2:.10f}, f(x2) = {f_b(x2):.3e}")
    if abs(x2 - x1) < tol:
        x0, x1 = x1, x2
        break
    x0, x1 = x1, x2
root_b = x1
print(f"Converged: root = {root_b:.8f} after {n+1} iterations\n")

#c
f_c = lambda x: np.exp(x) + np.sin(x) - 4

x0, x1 = 1.0, 2.0
print("Part (c): e^x + sin(x) - 4 = 0")
print(f"x0 = {x0}, x1 = {x1}")

for n in range(max_iter):
    f0, f1 = f_c(x0), f_c(x1)
    x2 = x1 - f1*(x1 - x0)/(f1 - f0)
    print(f"  n={n}: x2 = {x2:.10f}, f(x2) = {f_c(x2):.3e}")
    if abs(x2 - x1) < tol:
        x0, x1 = x1, x2
        break
    x0, x1 = x1, x2
root_c = x1
print(f"Converged: root = {root_c:.8f} after {n+1} iterations")
```

# CP.1.5.2, Sauer3 Langou
Use the Method of False Position to find the (single) solution of each of the following equations. Start from initial bracket $[1, 2]$.

a. $x^3 = 2x + 2$

b. $e^x + x = 7$

c. $e^x + \sin x = 4$

```python
tol = 0.5e-8
max_iter = 200

#a
f_a = lambda x: x**3 - 2*x - 2

a, b = 1.0, 2.0
fa, fb = f_a(a), f_a(b)
print("Part (a): x^3 - 2x - 2 = 0")
print(f"a = {a}, b = {b},  f(a) = {fa:.6f}, f(b) = {fb:.6f}")

for n in range(max_iter):
    c = b - fb*(b - a)/(fb - fa)
    fc = f_a(c)
    print(f"  n={n}: c = {c:.10f}, f(c) = {fc:.3e}")
    if abs(fc) < tol:
        break
    if fa*fc < 0:
        b, fb = c, fc
    else:
        a, fa = c, fc
root_a = c
print(f"Converged: root = {root_a:.8f} after {n+1} iterations\n")

#b
f_b = lambda x: np.exp(x) + x - 7

a, b = 1.0, 2.0
fa, fb = f_b(a), f_b(b)
print("Part (b): e^x + x - 7 = 0")
print(f"a = {a}, b = {b},  f(a) = {fa:.6f}, f(b) = {fb:.6f}")

for n in range(max_iter):
    c = b - fb*(b - a)/(fb - fa)
    fc = f_b(c)
    print(f"  n={n}: c = {c:.10f}, f(c) = {fc:.3e}")
    if abs(fc) < tol:
        break
    if fa*fc < 0:
        b, fb = c, fc
    else:
        a, fa = c, fc
root_b = c
print(f"Converged: root = {root_b:.8f} after {n+1} iterations\n")


#c
f_c = lambda x: np.exp(x) + np.sin(x) - 4

a, b = 1.0, 2.0
fa, fb = f_c(a), f_c(b)
print("Part (c): e^x + sin(x) - 4 = 0")
print(f"a = {a}, b = {b},  f(a) = {fa:.6f}, f(b) = {fb:.6f}")

for n in range(max_iter):
    c = b - fb*(b - a)/(fb - fa)
    fc = f_c(c)
    print(f"  n={n}: c = {c:.10f}, f(c) = {fc:.3e}")
    if abs(fc) < tol:
        break
    if fa*fc < 0:
        b, fb = c, fc
    else:
        a, fa = c, fc
root_c = c
print(f"Converged: root = {root_c:.8f} after {n+1} iterations")
```

# CP.1.5.3, Sauer3 Langou
Use Inverse Quadratic Interpolation to find the (single) solution of each of the following equations. Use initial guesses $x_0 = 1$, $x_1 = 2$, $x_2 = 0$, and update by retaining the three most recent iterates.

a. $x^3 = 2x + 2$

b. $e^x + x = 7$

c. $e^x + \sin x = 4$

```python
tol = 0.5e-8
max_iter = 100

#a
f_a = lambda x: x**3 - 2*x - 2

x0, x1, x2 = 1.0, 2.0, 0.0
print("Part (a): x^3 - 2x - 2 = 0")

for n in range(max_iter):
    f0, f1, f2 = f_a(x0), f_a(x1), f_a(x2)
    x3 = ( f1*f2/((f0-f1)*(f0-f2)) * x0
         + f0*f2/((f1-f0)*(f1-f2)) * x1
         + f0*f1/((f2-f0)*(f2-f1)) * x2 )
    if abs(x3 - x2) < tol:
        x0, x1, x2 = x1, x2, x3
        break
    x0, x1, x2 = x1, x2, x3
root_a = x2
print(f"Converged: root = {root_a:.8f} after {n+1} iterations\n")

#b
f_b = lambda x: np.exp(x) + x - 7

x0, x1, x2 = 1.0, 2.0, 0.0
print("Part (b): e^x + x - 7 = 0")

for n in range(max_iter):
    f0, f1, f2 = f_b(x0), f_b(x1), f_b(x2)
    x3 = ( f1*f2/((f0-f1)*(f0-f2)) * x0
         + f0*f2/((f1-f0)*(f1-f2)) * x1
         + f0*f1/((f2-f0)*(f2-f1)) * x2 )
    if abs(x3 - x2) < tol:
        x0, x1, x2 = x1, x2, x3
        break
    x0, x1, x2 = x1, x2, x3
root_b = x2
print(f"Converged: root = {root_b:.8f} after {n+1} iterations\n")

#c
f_c = lambda x: np.exp(x) + np.sin(x) - 4

x0, x1, x2 = 1.0, 2.0, 0.0
print("Part (c): e^x + sin(x) - 4 = 0")

for n in range(max_iter):
    f0, f1, f2 = f_c(x0), f_c(x1), f_c(x2)
    x3 = ( f1*f2/((f0-f1)*(f0-f2)) * x0
         + f0*f2/((f1-f0)*(f1-f2)) * x1
         + f0*f1/((f2-f0)*(f2-f1)) * x2 )
    if abs(x3 - x2) < tol:
        x0, x1, x2 = x1, x2, x3
        break
    x0, x1, x2 = x1, x2, x3
root_c = x2
print(f"Converged: root = {root_c:.8f} after {n+1} iterations")
```
