## EX.0.1.6 (Sauer)

Explain how to evaluate the polynomial below for a given input $x$, using as few operations as possible. You are **not** allowed to use the power operator (e.g. no `x**5`). You may only use multiplications and additions. How many multiplications and how many additions are required?

**a.** $p(x) = a_0 + a_5 x^5 + a_{10} x^{10} + a_{15} x^{15}$

**b.** $p(x) = a_7 x^7 + a_{12} x^{12} + a_{17} x^{17} + a_{22} x^{22} + a_{27} x^{27}$

```python
#a

a0, a5, a10, a15 = 1, 2, 3, 4
x = 2

x2 = x * x
x4 = x2 * x2
y = x4 * x

p = a0 + y * (a5 + y * (a10 + a15 * y))
print("p(x) =", p)
```

```python
#b

a7, a12, a17, a22, a27 = 1, 2, 3, 4, 5
x = 2

x2 = x * x
x4 = x2 * x2
y = x4 * x  # y = x^5

p = x2 * y * (a7 + y * (a12 + y * (a17 + y * (a22 + a27 * y))))
print("p(x) =", p)
```
